```mojo
%%python
digits = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'seven':7, 'eight':8, 'nine':9, 'zero':0}


'''
oneight

    .
o   t   f     s   e   n    z
n   wh  oi    ie  i   i    e
e   or  uv    xv  g   n    r
     e  re     e  h   e    o
     e         n  t

'''

class PrefixTree:
        
    def __init__(self, val, parent = None):
        self.value = val
        self.parent = parent
        self.children = {}
        
    def add_word(self, word):
        if word[0] in self.children.keys():
            self.children[word[0]].add_word(word[1:])
        else:
            self.children[word[0]] = PrefixTree(word[0], self)
            if len(word) > 1:
                self.children[word[0]].add_word(word[1:])
    
    def print_words(self, word=''):
        word += self.value
        if len(self.children) == 0:
            print(word)
            return
        for c in self.children.values():
            c.print_words(word)
    
    def prefix(self):
        s = self.value
        t = self
        while t.parent:
            s = t.parent.value + s
            t = t.parent
        return s

tree = PrefixTree('')
for d in digits.keys():
    tree.add_word(d)
tree.print_words()

print('\n', tree.children['t'].children['w'].children['o'].prefix())
```
